```python
df['Common_Symtoms'] = df['Common_Symtoms'].apply(lambda x: tuple(x) if isinstance(x, list) else x)
print("Duplicates found:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
import numpy as np

#  outliers using IQR
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    print(f"{col}: {len(outliers)} outliers")
    
    # cap them (Winsorization)
    df[col] = np.where(df[col] > Q3 + 1.5 * IQR, Q3 + 1.5 * IQR,
                       np.where(df[col] < Q1 - 1.5 * IQR, Q1 - 1.5 * IQR, df[col]))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols_to_check = [
    'total_cases_per_million',
    'new_cases_per_million',
    'total_deaths_per_million',
    'new_deaths_per_million'
]

for col in cols_to_check:
    plt.figure(figsize=(8, 1))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

# Apply to all four numeric columns
for col in cols_to_check:
    df = remove_outliers_iqr(df, col)
```

```python
# Decision Tree, KNN & Naive Bayes

import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report
)

# Dictionary of models
models = {
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Naive Bayes": GaussianNB()
}

# Loop through each model
for name, model in models.items():
    print(f"\n Model: {name}")
    
    # Create pipeline with imputer and classifier
    pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), model)
    
    # Fit on training data
    pipeline.fit(X_train, y_train)
    
    # Predict on test data
    y_pred = pipeline.predict(X_test)
    
    # Evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    print(f"   Accuracy : {accuracy:.4f}")
    print(f"   Precision: {precision:.4f}")
    print(f"   Recall   : {recall:.4f}")
    print(f"   F1 Score : {f1:.4f}")
    
    # Display full classification report
    print(f"\n Classification Report:\n{classification_report(y_test, y_pred)}")
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=pipeline.classes_)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

```python
# ROC-AUC Curve for both train and test dataset

from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

def evaluate_model_roc(model, X_train, y_train, X_test, y_test, model_name="Model"):
    """
    Fits a pipeline (with imputation) and plots ROC curve for the given classifier.
    
    Parameters:
    - model: A scikit-learn classifier (e.g., KNeighborsClassifier())
    - X_train, y_train: Training data
    - X_test, y_test: Testing data
    - model_name: String label for the plot title and legend
    
    Returns:
    - auc_train: AUC score on training data
    - auc_test: AUC score on test data
    
    """

    # Build pipeline
    pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), model)
    pipeline.fit(X_train, y_train)

    # Predict probabilities for positive class
    y_train_prob = pipeline.predict_proba(X_train)[:, 1]
    y_test_prob = pipeline.predict_proba(X_test)[:, 1]

    # ROC curve and AUC
    fpr_train, tpr_train, _ = roc_curve(y_train, y_train_prob)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)
    auc_train = auc(fpr_train, tpr_train)
    auc_test = auc(fpr_test, tpr_test)

    # Plot
    plt.figure(figsize=(6, 5))
    plt.plot(fpr_train, tpr_train, label=f'Train AUC = {auc_train:.4f}')
    plt.plot(fpr_test, tpr_test, label=f'Test AUC = {auc_test:.4f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f'ROC Curve - {model_name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid(True)
    plt.show()

    return auc_train, auc_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression

# Evaluate KNN
evaluate_model_roc(KNeighborsClassifier(), X_train, y_train, X_test, y_test, model_name="KNN")

# Evaluate Decision Tree
evaluate_model_roc(DecisionTreeClassifier(random_state=42), X_train, y_train, X_test, y_test, model_name="Decision Tree")

# Evaluate Naive Bayes
evaluate_model_roc(GaussianNB(), X_train, y_train, X_test, y_test, model_name="Naive Bayes")

# Evaluate Logistic Regression
evaluate_model_roc(LogisticRegression(max_iter=1000), X_train, y_train, X_test, y_test, model_name="Logistic Regression")
```

```python
import os
import joblib
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

# Setup Paths
BASE_DIR = os.path.abspath("..")
MODEL_DIR = os.path.join(BASE_DIR, 'models')
os.makedirs(MODEL_DIR, exist_ok=True)

# Load Dataset
data_path = os.path.join(BASE_DIR, 'data', 'Cleaned Data of Covid-19.csv')
df = pd.read_csv(data_path)

# Regression Setup
regression_features = [
    'total_cases_per_million','fever', 'dry cough', 'fatigue',
    'loss of taste', 'loss of or smell', 'difficulty breathing',
    'sore throat', 'headache', 'muscle aches', 'chills', 'diarrhea',
    'runny nose', 'nausea'
]

X_reg = df[regression_features]
y_reg = df['total_deaths_per_million']

print("Regression Features:")
print(X_reg.columns)
print("Regression Target Sample:")
print(y_reg.head())

# Classification Setup
X_clf = df.drop(columns=['high_death_risk'])

drop_cols = [
    'total_cases_per_million',
    'total_deaths_per_million',
    'month',
    'day',
    'location_Bangladesh',
    'location_India',
    'location_Pakistan'
]

X_clf = X_clf.drop(columns=drop_cols)
y_clf = df['high_death_risk']

print("\nClassification Features:")
print(X_clf.columns)
print("Classification Target Sample:")
print(y_clf.head())

# Train Regression Models
lr_model = LinearRegression()
lr_model.fit(X_reg, y_reg)
joblib.dump(lr_model, os.path.join(MODEL_DIR, 'linear_regression_model.pkl'))
print("Saved: Linear Regression")

poly_pipeline = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_pipeline.fit(X_reg, y_reg)
joblib.dump(poly_pipeline, os.path.join(MODEL_DIR, 'polynomial_regression_model.pkl'))
print("Saved: Polynomial Regression")

# Train Classification Models
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

rf_model = RandomForestClassifier()
rf_model.fit(X_train_clf, y_train_clf)
joblib.dump(rf_model, os.path.join(MODEL_DIR, 'random_forest_model.pkl'))
print("Saved: Random Forest")

log_reg_pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), LogisticRegression(max_iter=1000))
log_reg_pipeline.fit(X_train_clf, y_train_clf)
joblib.dump(log_reg_pipeline, os.path.join(MODEL_DIR, 'logistic_regression_model.pkl'))
print("Saved: Logistic Regression")

knn_pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), KNeighborsClassifier())
knn_pipeline.fit(X_train_clf, y_train_clf)
joblib.dump(knn_pipeline, os.path.join(MODEL_DIR, 'knn_model.pkl'))
print("Saved: KNN")

dt_pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), DecisionTreeClassifier(random_state=42))
dt_pipeline.fit(X_train_clf, y_train_clf)
joblib.dump(dt_pipeline, os.path.join(MODEL_DIR, 'decision_tree_model.pkl'))
print("Saved: Decision Tree")

nb_pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), GaussianNB())
nb_pipeline.fit(X_train_clf, y_train_clf)
joblib.dump(nb_pipeline, os.path.join(MODEL_DIR, 'naive_bayes_model.pkl'))
print("Saved: Naive Bayes")

print("\n All models trained and saved successfully!")
```
